```sql
USE CATALOG airbnb_lab3;

CREATE OR REPLACE VIEW airbnb_gold.v_city_overview AS
SELECT city, yyyymm, total_listings, avg_nightly_price, avg_occupancy_rate, avg_review_score
FROM airbnb_gold.gold_city_month_metrics;

CREATE OR REPLACE VIEW airbnb_gold.v_neighbourhood AS
SELECT city, yyyymm, neighbourhood, listing_count, avg_price, avg_occupancy_rate
FROM airbnb_gold.gold_neighbourhood_metrics;

CREATE OR REPLACE VIEW airbnb_gold.v_roomtype_share AS
SELECT city, yyyymm, room_type, share
FROM airbnb_gold.gold_roomtype_share;

CREATE OR REPLACE VIEW airbnb_gold.v_reviews_monthly AS
SELECT city, yyyymm, review_count
FROM airbnb_gold.gold_reviews_monthly;

CREATE OR REPLACE VIEW airbnb_gold.v_cross_city AS
SELECT * FROM airbnb_gold.gold_cross_city_compare;
```
